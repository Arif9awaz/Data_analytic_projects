```python
#It looks like the num_photos column has 0 for every column.
#We'll drop this column, plus the other two we noted as mostly one value.

autos = autos.drop(["num_photos", "seller", "offer_type"], axis=1)
```

```python
#KeyError: "['num_photos', 'seller', 'offer_type'] not found in axis" because we have droped the columns
# let's check them in the dataframe
autos.head()
```

```python
# here in price column $ and , will be considered as  characters so let's remove it
autos['price']=(autos['price'].str.replace('$','')
                .str.replace(',','')
                .astype(int))
autos['price'].head()
```

```python
#removing km and , from odometer column

autos["odometer"] = (autos["odometer"]
 .str.replace("km","")
 .str.replace(",","")
 .astype(int)
                    )

                     
```

```python
#Given that eBay is an auction site, there could legitimately be items where the opening bid is $1 .
#We will keep the $1 items, but remove anything above $350,000 ,
 #since it seems that prices increase steadily to that number and then jump up to less realistic numbers
autos= autos[autos['price'].between(1,351000)]
autos['price'].sort_index(ascending=True).head(100)
    
```

```python
# Many ways to select rows in a dataframe that fall within a value range for a column.
# Using `Series.between()` is one way.
autos = autos[autos["registration_year"].between(1900,2016)]
autos["registration_year"].value_counts(normalize=True).head(10)
```

```python
# Exploring Price by Brand
autos['brand'].value_counts(normalize=True)
```

```python
brand_counts = autos["brand"].value_counts(normalize=True)
common_brands = brand_counts[brand_counts > .05].index
print(common_brands)
```

```python
brand_mean_price={}
for brand in common_brands:
    only_brand=autos[autos['brand']==brand]
    mean_price=only_brand['price'].mean()
    brand_mean_price[brand]=int(mean_price)
    
```

```python
brand_mean_price
```

```python
#Of the top 5 brands, there is a distinct price gap:
#Audi, BMW and Mercedes Benz are more expensive
#Ford and Opel are less expensive
#Volkswagen is in between - this may explain its popularity, it may be a 'best of 'both worlds' option.

```

```python
# putting the brand's data into dataframe
brand_df=pd.Series(brand_mean_price)
pd.DataFrame(brand_df,columns=['mean_price'])
```
