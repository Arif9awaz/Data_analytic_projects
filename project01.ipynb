```python

q1 = """
select g.name,count(t.track_id) as total_sales from 
genre g
inner join 
track t
on
t.genre_id = g.genre_id

inner join 
invoice_line il
on
t.track_id = il.track_id

inner join invoice i
on
i.invoice_id = il.invoice_id

inner join customer c
on
c.customer_id = i.customer_id
where c.country="USA"
group by g.name
order by total_sales desc



"""

pd.read_sql(q1, conn)
```

```python
q1="""with final as(select g.name,count(t.track_id) as total_sales from 
genre g
inner join 
track t
on
t.genre_id = g.genre_id

inner join 
invoice_line il
on
t.track_id = il.track_id

inner join invoice i
on
i.invoice_id = il.invoice_id

inner join customer c
on
c.customer_id = i.customer_id
where c.country="USA"
group by g.name
order by total_sales desc),
tsales as(select sum("total_sales") ts from final)
select name,total_sales,
cast("total_sales" as float)/(select ts from tsales) pct_sales from final
"""
query=pd.read_sql(q1,conn)
```

```python
result=pd.read_sql(q1,conn)
result.set_index("name",inplace = True, drop = True)
#label=result["name"]

ts=result["total_sales"]
pcts=result["pct_sales"]
annt=(pcts*100).astype(int).astype(str)+"%"
for x,y,txt in zip(ts, range(17),annt):
    plt.annotate(txt, (x  + 10, y -0.15))

ax=result.total_sales.plot.barh(title = "Top Selling Genres in the USA by total_sales in percentage",
                                         colormap = plt.cm.Accent,
                                         xlim = (0,600))

ax.tick_params(which="major",left ="off",color="red")
ax.set_frame_on(False)



#ax.spines['top'].set_visible("off")

plt.show()
```

```python
#2 scenario analyzing sales by country
q="""select country,t_customers,total_sales,average_order,customer_lifetime_value from(select country,count(distinct c.customer_id) t_customers,  sum(unit_price) total_sales, 
            sum(il.unit_price)/count(distinct i.invoice_id) average_order,
            sum(il.unit_price)/count(distinct c.customer_id) customer_lifetime_value from customer c 
inner join
invoice i on 
c.customer_id=i.customer_id
inner join
invoice_line il on
i.invoice_id=il.invoice_id

group by c.country
order by t_customers desc) where t_customers>1
"""
bbc=pd.read_sql(q,conn)


```

```python
bbc=pd.read_sql(q,conn)
bbc.set_index("country", inplace = True, drop = True)
fig, ax = plt.subplots(1,2,figsize = (9,10))
ax1, ax2 = ax.flatten()

# top left
Sales_breakdown=bbc["total_sales"]
Sales_breakdown.plot.pie(ax=ax1, title = "Sales Breakdown by Country,\nNumber of Customers",
                         colormap=plt.cm.Accent, startangle = 45,
                         counterclock = False, fontsize = 8)
#plt.tight_layout()

#plt.show()
# top right
tc_ts=['t_customers','total_sales']
fcs=bbc[tc_ts]
#fcs.index.name = ""







    
fcs.plot.bar(ax = ax2, colormap = plt.cm.Set1,
                          title ="Pct Customers vs Sales")
ax2.tick_params(top = "off", right = "off", left = "off", bottom = "off")
ax2.spines["top"].set_visible(False)
ax2.spines["right"].set_visible(False)
ax2.tick_params(top=False, right=False, left=False, bottom=False)

#ax3.legend(loc = "center")
plt.show()
```
